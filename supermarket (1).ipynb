```python
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd

# -----------------------------
# LOAD DATA (change path if needed)
# -----------------------------
df = pd.read_excel("C:/Users/abira/Downloads/Superstore.xlsx")

# -----------------------------
# KPI VALUES
# -----------------------------
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].nunique()

# -----------------------------
# CHARTS
# -----------------------------

# Sales by Category
fig_sales_category = px.bar(df.groupby("Category", as_index=False)["Sales"].sum(),
                            x="Category", y="Sales",
                            title="Sales by Category")

# Profit by Category
fig_profit_category = px.bar(df.groupby("Category", as_index=False)["Profit"].sum(),
                             x="Category", y="Profit",
                             title="Profit by Category")

# Sales by Region
fig_sales_region = px.bar(df.groupby("Region", as_index=False)["Sales"].sum(),
                          x="Region", y="Sales",
                          title="Sales by Region")

# Profit by Region
fig_profit_region = px.bar(df.groupby("Region", as_index=False)["Profit"].sum(),
                           x="Region", y="Profit",
                           title="Profit by Region")

# Year vs Sales
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year

fig_year_sales = px.line(df.groupby("Year", as_index=False)["Sales"].sum(),
                         x="Year", y="Sales",
                         title="Year vs Sales")

# Top 10 Products by Sales
top_products = df.groupby("Product Name", as_index=False)["Sales"].sum()
top_products = top_products.sort_values("Sales", ascending=False).head(10)

fig_top_products = px.bar(top_products,
                          x="Sales", y="Product Name",
                          orientation="h",
                          title="Top 10 Products by Sales")

# -----------------------------
# DASH APP
# -----------------------------
app = Dash(__name__)

app.layout = html.Div([

    # Title
    html.H1("📊 Superstore Sales Dashboard", style={"textAlign": "center"}),

    # KPI Cards
    html.Div([
        html.Div(f"Total Sales: {round(total_sales,2)}"),
        html.Div(f"Total Profit: {round(total_profit,2)}"),
        html.Div(f"Total Orders: {total_orders}")
    ], style={"display": "flex", "justifyContent": "space-around"}),

    html.Hr(),

    # Charts Row 1
    html.Div([
        dcc.Graph(figure=fig_sales_category),
        dcc.Graph(figure=fig_profit_category)
    ], style={"display": "flex"}),

    # Charts Row 2
    html.Div([
        dcc.Graph(figure=fig_sales_region),
        dcc.Graph(figure=fig_profit_region)
    ], style={"display": "flex"}),

    # Year trend
    dcc.Graph(figure=fig_year_sales),

    # Top products
    dcc.Graph(figure=fig_top_products)

])

# -----------------------------
# RUN SERVER
# -----------------------------
if __name__ == "__main__":
    app.run(debug=True)
```
